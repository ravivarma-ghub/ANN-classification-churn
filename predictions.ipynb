```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
#Loading the models
model=load_model('model.h5')

#load the encoder and scaler
with open('onehot_encoder_geo.pkl','rb') as file:
    label_encoder_geo=pickle.load(file)
with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder=pickle.load(file)
```

```python
import pandas as pd

# Example input data
input_data = {
    'CreditScore': [600],
    'Geography': ['France'],
    'Gender': ['Male'],
    'Age': [40],
    'Tenure': [30],
    'Balance': [60000],
    'NumOfProducts': [2],
    'HasCrCard': [1],
    'IsActiveMember': [1],
    'EstimatedSalary': [50000]
}

# Convert input data dictionary to DataFrame
input_df = pd.DataFrame(input_data)
```

```python
geo_encoded = label_encoder_geo.transform(input_df[['Geography']]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded, columns=label_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_data=pd.concat([input_df.drop('Geography', axis=1), geo_encoded_df], axis=1)
```

```python
input_data
```

```python
input_data['Gender'] = label_encoder.transform(input_data['Gender'])
input_data
```

```python
input_scaled = scaler.transform(input_data)
```

```python
#Predictions
model.predict(input_scaled)
```
